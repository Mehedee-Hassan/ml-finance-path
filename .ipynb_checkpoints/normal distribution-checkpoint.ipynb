```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
data = np.linspace(-3,3,num =1000)
```

```python
mean = 0 
std = 1
```

```python
pdf = stats.norm.pdf(data,mean,std)
```

```python
plt.plot(data, pdf,'-', color='black', lw=2)
plt.axvline(mean,color='black', linestyle='--')
plt.grid()
plt.show()
```
